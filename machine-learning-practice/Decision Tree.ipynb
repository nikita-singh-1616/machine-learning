```python
# for classification problems sometimes the data is so diversely scattered that a lines cannot classify the data into categories 
# in such cases we use decision tree
import pandas as pd
employee_salary_data = pd.read_csv('salaries.csv')
employee_salary_data.head()
```

```python
# since out model only reads numeric data we encode this data 
from sklearn.preprocessing import LabelEncoder
company_encoder = LabelEncoder()
job_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
employee_salary_data['encoded_company'] = company_encoder.fit_transform(employee_salary_data.company)
employee_salary_data['encoded_job'] = company_encoder.fit_transform(employee_salary_data.job)
employee_salary_data['encoded_degree'] = company_encoder.fit_transform(employee_salary_data.degree)

employee_salary_data
```

```python
employee_salary_data = employee_salary_data.drop(['company','job','degree'],axis='columns')
employee_salary_data
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
model = tree.DecisionTreeClassifier()
x = employee_salary_data.drop(['salary_more_then_100k'],axis='columns')
y = employee_salary_data['salary_more_then_100k']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test),y_test
```
